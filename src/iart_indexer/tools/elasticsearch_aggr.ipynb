```python
import json
from elasticsearch import Elasticsearch, exceptions
es = Elasticsearch([{"host": 'localhost', "port": 9200}], timeout=30)
```

```python
field_name = "institution"
size = 2
```

```python
body = {
    "aggs": {
        "meta_nested": {
            "nested": {"path": "meta"},
            "aggs": {
                "meta_name_filter": {
                    "filter": {"term": {"meta.name": field_name}},
                    "aggs": {
                        "meta_name_filter_aggr": {"terms": {"size": size, "field": "meta.value_str.keyword"}}
                    },
                }
            },
        }
    }
}
result =es.search(index='iart_all_210223', body=body, size=0)
print(json.dumps(result, indent=2))
```
